#Importing the dataset

```python
dataset = pd.read_csv('insurance.csv')
dataset
```

Now  the dataset needs to be split into independent and dependent values which are features and target respectively. Target values are dependent on feature values as the targets need to be predicted with the help of feature values.

```python
feature = dataset[['age','sex','bmi','children','smoker','region']]
target = dataset[['charges']]
```

```python
feature
```

```python
target
```

#Taking care of  missing data

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy = "mean")
imputer = imputer.fit(feature[['age','bmi','children']]) 
```

```python
feature[['age','bmi','children']]= imputer.transform(feature[['age','bmi','children']])
```

```python
feature
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy = "mean")
imputer = imputer.fit(target[['charges']]) 
target[['charges']]= imputer.transform(target[['charges']])
```

```python
target
```

#Encoding categorical data using one hot encoding 


```python
feature['sex'].unique()
feature['sex'].value_counts()
```

```python
feature['smoker'].unique()
feature['smoker'].value_counts()
```

```python
feature['region'].unique()
feature['region'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
encoder=OneHotEncoder(sparse=False)
encoded_labels = pd.DataFrame(encoder.fit_transform(feature[['sex','smoker','region']]))
```

```python
encoded_labels.columns = encoder.get_feature_names(['sex','smoker','region'])
dataset= pd.concat([feature, encoded_labels], axis=1)
dataset
```

```python
dataset.columns
```

```python
new_feature = dataset[['age',	'bmi',	'children','sex_female','sex_male','smoker_yes','smoker_no','region_northeast','region_northwest','region_southeast','region_southwest']]
```

#Spliting the Dataset into Training Set and Test Set



```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(new_feature,target,test_size=0.2,random_state=0)
# random_state = 0 is select to get the same result
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_train
```

```python
X_test
```

#Multivariant Linear Regression

```python
# Fitting Multiple Linear Regression Model to Training Set
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train,y_train)

# Predicting the Test Set Results
y_pred = regressor.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error

# MAE

mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```

```python
dataset
```

```python
from sklearn.preprocessing import StandardScaler
X_sc = StandardScaler()
y_sc = StandardScaler()
X_train_new = X_sc.fit_transform(X_train[['age',	'bmi',	'children','sex_female','sex_male','smoker_yes','smoker_no','region_northeast','region_northwest','region_southeast','region_southwest']])
X_test_new = X_sc.transform(X_test[['age',	'bmi',	'children','sex_female','sex_male','smoker_yes','smoker_no','region_northeast','region_northwest','region_southeast','region_southwest']])
y_train_new = y_sc.fit_transform(y_train[['charges']])
y_test_new = y_sc.transform(y_test[['charges']])
```

#Support Vector Regression

```python
# Fitting SVR to the dataset
from sklearn.svm import SVR 

regressor = SVR(kernel = 'linear')
regressor.fit(X_train_new,y_train_new)
```

```python
y_pred = regressor.predict(X_test_new)
```

```python
r2_score(y_test_new, y_pred)
```

#Decision Tree Regression

```python
# Fitting the decision tree regression with the dataset
from sklearn.tree import DecisionTreeRegressor


regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X_train_new,y_train_new)
```

```python
y_pred = regressor.predict(X_test_new)
```

```python
r2_score(y_test_new, y_pred)
```

#Random Forest Regression

```python
# Fitting the Random Forest Regression with the dataset
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators = 10,random_state = 0) # n estiamator is the number of decision trees
regressor.fit(X_train_new,y_train_new) 
```

```python
y_pred = regressor.predict(X_test_new)
r2_score(y_test_new, y_pred)
```
